## Aquisição de daos

```python
data = pd.read_csv('caso.csv')
```

## Pegando os dados do piaui

```python
data_piaui = data[data['state'] == 'PI']
```

```python
cidades = data_piaui[(data_piaui['is_last'] == True)]
```

```python
len(cidades)
```

```python
cidades.head()
```

```python
cidades.info()
```

```python
cidades_t = cidades.dropna(subset=["city"])
cidades_t.info()
```

```python
cidades_target = cidades_t["deaths"].copy()
cidades_t = cidades_t.drop('deaths', axis=1) 
cidades_t
```

Dados faltantes: cidades_t = cidades.dropna(subset=[""]) 


```python
cidades_float = cidades_t.drop(['date', 'state', 'city', 'place_type', 'is_last'], axis=1)
cidades_float
```

## Dividando os dados em treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(cidades_float, cidades_target, test_size = 0.2,random_state=1, shuffle=True)
```

```python
#z = (x - u) / s
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Salvando os conjuntos em pickle

```python
pickle.dump(X_train, open('X_train.pickle', 'wb'))
pickle.dump(X_test, open('X_test.pickle', 'wb'))
pickle.dump(y_train, open('y_train.pickle', 'wb'))
pickle.dump(y_test, open('y_test.pickle', 'wb'))
```
